```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

import warnings
warnings.filterwarnings("ignore")
```

```python
mae = list()
mse = list()
rmse = list()
r2 = list()
mape = list()

x = np.arange(-1000, 1001, 1)
for _x in x:
    mae.append(mean_absolute_error([_x], [0]))
    mse.append(mean_squared_error([_x], [0]))
    rmse.append(np.sqrt(mean_squared_error([_x], [0])))
    mape.append(mean_absolute_percentage_error([_x], [0]))

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.plot(x, mae)
plt.title('MAE')
plt.subplot(2, 2, 2)
plt.plot(x, mse)
plt.title('MSE')
plt.subplot(2, 2, 3)
plt.plot(x, rmse)
plt.title('RMSE')
plt.subplot(2, 2, 4)
plt.plot(x, mape)
plt.title('MAPE')
plt.show()
```
